# Step 2 (Join and Clean Scraped Data)

## 1. Import all dependecies here

```python
import pandas as pd 
```

```python
df_salary=pd.read_csv("/Users/main/Desktop/dsci510/project/salary_23_24",index_col='Unnamed: 0')
df_stats=pd.read_csv("/Users/main/Desktop/dsci510/project/stats_23_24", index_col='Unnamed: 0')
```

```python
df_salary.columns = df_salary.columns.str.replace(df_salary.columns[0], 'Player')
df_salary.columns = df_salary.columns.str.replace(df_salary.columns[1], 'salary')
df_salary
```

```python
#salary
#first name:a[0], lastname:a[1]
a = df_salary['Player'].values.tolist()
for i in range(len(a)): 
    a[i]= a[i].lower()
    a[i] = a[i].split(" ")
    if(len(a[i]))!=2:
        a[i]=a[i][:2]
        print(a[i])
```

```python
df_salary["first_name"]=[first[0] for first in a]
df_salary["last_name"]=[last[1] for last in a]
df_salary
```

```python
#stats
#first name:b[0], lastname:b[1]
b = df_stats['Player'].values.tolist()
for i in range(len(b)):
    b[i]= b[i].lower()
    b[i] = b[i].split(" ")
    if(len(b[i]))!=2:
        print(b[i])
        b[i]=b[i][:2]
        print(b[i])
```

```python
df_stats["first_name"]=[first[0] for first in b]
df_stats["last_name"]=[last[1] for last in b]
df_stats
```

```python
df_salary.columns = df_salary.columns.str.replace(df_salary.columns[1], 'salary')
df_salary
```

```python
merged_df = pd.merge(df_salary, df_stats, on=['first_name', 'last_name'], how='inner')
print(len(df_salary))
print(len(df_stats))
print(len(merged_df))
```

```python
merged_df['salary'] = merged_df['salary'].str.replace('$', '').str.replace(',', '').astype(float)
```
